```python
import matplotlib.pyplot as plt  # for drawing graphs

# Define the slope and intercept
m = 2        # This is the slope (the line goes up 2 units for every 1 unit across)
b = 3        # This is the y-intercept (where the line cuts the y-axis)

# Create x values (from -10 to 10)
x = list(range(-10, 11))         # x will be a list of numbers from -10 to 10

# Create y values using the formula y = mx + b
y = [m * num + b for num in x]   # y = 2x + 3 for each value of x

# Plot the line
plt.plot(x, y, label='y = 2x + 3')  # draw the line and label it
plt.xlabel('x')                    # label for x-axis
plt.ylabel('y')                    # label for y-axis
plt.title('Line with slope 2 and intercept 3')  # graph title
plt.axhline(0, color='black')      # x-axis line
plt.axvline(0, color='black')      # y-axis line
plt.grid(True)                     # turn on grid for better visibility
plt.legend()                       # show the label in the plot
plt.show()                         # show the graph
```

```python
import matplotlib.pyplot as plt  # for drawing graphs

# Define slope and intercept
m = -1       # slope is -1 (line goes down 1 unit for every 1 unit right)
b = -2       # intercept is -2 (the line crosses y-axis at -2)

# Create x values
x = list(range(-10, 11))              # x goes from -10 to 10

# Create y values using the formula y = mx + b
y = [m * num + b for num in x]        # y = -1x - 2

# Plot the line
plt.plot(x, y, label='y = -x - 2')    # draw the line and add a label
plt.xlabel('x')                       # label for x-axis
plt.ylabel('y')                       # label for y-axis
plt.title('Line with slope -1 and intercept -2')  # title of the graph
plt.axhline(0, color='black')         # x-axis
plt.axvline(0, color='black')         # y-axis
plt.grid(True)                        # show grid
plt.legend()                          # show the label in the plot
plt.show()                            # display the graph
```
